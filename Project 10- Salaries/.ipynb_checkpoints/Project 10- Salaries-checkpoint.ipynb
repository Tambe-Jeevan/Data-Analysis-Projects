## Handling missing values

```python
# Since Notes and status have most of the values as NULL, we drop it from column
df.drop(columns=['Notes','Status','Agency'],axis = 1,inplace=True)
```

```python
df.head(10)
```

```python
df['BasePay'] = df['BasePay'].replace('Not Provided','0')
df['BasePay'] = df.BasePay.astype('float')
```

```python
basepay_mean = df.BasePay.mean()
basepay_mean
```

```python
df['BasePay'] = df['BasePay'].replace(np.nan,basepay_mean)
```

```python
df['OvertimePay'].unique()
```

```python
df['OvertimePay'] = df['OvertimePay'].replace('Not Provided','0')
df['OvertimePay'] = df.OvertimePay.astype('float')

overtimepay_mean = df.OvertimePay.mean()
df['OvertimePay'] = df['OvertimePay'].replace(np.nan, overtimepay_mean)
```

```python
df.isnull().sum()
```

```python
df.Benefits.unique()
```

```python
df['Benefits'] = df['Benefits'].replace('Not Provided','0')
df['Benefits'] = df.Benefits.astype('float')

benefit_mean = df.Benefits.mean()
df['Benefits'] = df['Benefits'].replace(np.nan, benefit_mean)
```

```python
df['OtherPay'] = df['OtherPay'].replace('Not Provided','0')
df['OtherPay'] = df.OtherPay.astype('float')

otherpay_mean = df.OtherPay.mean()
df['OtherPay'] = df['OtherPay'].replace(np.nan, otherpay_mean)
```
