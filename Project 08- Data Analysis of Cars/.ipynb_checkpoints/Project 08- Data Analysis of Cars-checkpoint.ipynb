## Data cleaning

## 1. Instruction (For Data Cleaning): Find all Null values in the dataset. If there is any null value in any column, then fill it with mean of the column

```python
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
# column names present in column are float datatypes, so their mean can be easily calculated.
column=['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway',
       'Weight', 'Wheelbase', 'Length']

for col in column:
    df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# the remaining column are object type, so their null values are removed without calculating any mean or whatever.
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df
# table size reduce to 428 x 15 compared to 432 x 15 previously
```
